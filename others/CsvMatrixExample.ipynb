```python
import numpy as np
```

```python
# load matrix from csv file
a = np.loadtxt("data/covtype",delimiter=",")
```

```python
a.shape, a.dtype
```

```python
# create x vector containing 1's
x = np.array([1]*a.shape[1])
```

```python
# elements are int (use [1.0]*54 to get floats)
x.shape, x.dtype
```

```python
# right multiplication 
y = a @ x
```

```python
# entries are now float
y.shape, y.dtype
```

```python
y[:20]
```

```python
# left multiplication
z = y @ a
```

```python
# orrible display format
z[:20]
```

```python
# to get a better representation we need to change the print options
np.get_printoptions()
```

```python
# our entries are ints: no decimals and don't use scientific notation
float_formatter = "{:.0f}".format
np.set_printoptions(formatter={'float_kind':float_formatter})
```

```python
# now we can see all the digits
# these are exactly the values in the readme file!
z[:20]
```
