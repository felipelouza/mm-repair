```python
# eigenvalue computation for a set of matrices 
import numpy as np
from scipy.linalg import eig, eigh
from scipy.sparse.linalg import eigs, eigsh
```

```python
# matrix names
names = ['covtype','census','optical']
prefix = "data/"
```

```python
for n in names:
    # load matrix
    m =  np.loadtxt(prefix + n,delimiter=",")
    print("### Matrix:", n, "  shape: %dx%d" % m.shape )
    eival, eivec = eigsh(m.T@m, 1, which='LM')
    print(eivec[:5])
    print(" Largest eigenvalue:", eival)
```
